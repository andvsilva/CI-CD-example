# Join (or merge) DataFrames using SQL Queries

```python
query = '''
SELECT m.date, m.beef, m.veal, m.pork, b.births
FROM meat AS m
INNER JOIN
birth AS b
ON m.date = b.date;
'''

mysql_query = psql.sqldf(query)
mysql_query
```
